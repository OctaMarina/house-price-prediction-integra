# E. Analiza distributiilor

# Distributia preturilor (variabila SalePrice)

```python

plt.hist(df_subset['SalePrice'],bins=30, color="cyan", edgecolor="grey")
plt.title('Distributia preturilor(SalePrice)')
plt.xlabel('SalePrice')
plt.ylabel('Frecventa')
plt.show()

```

# Forma distributiei si valori extreme

```python
df_subset.hist(figsize=(12, 10), bins=30, color="cyan")
plt.tight_layout()
plt.show()
```

# Boxplot-uri pentru variabilele numerice pentru a identifica outlierii si dispersia datelor

```python
numeric_cols = df_subset.select_dtypes(include='number').columns

for col in numeric_cols:
    df_subset.boxplot(column=col, color="b")
    plt.title(f"Boxplot pentru {col}")
    plt.show()
```

# Analiza distributiei variabilelor categorice folosind diagrame de tip barplot
# (seaborn - functia countplot) si vizualizarea frecventei fiecărei categorii

```python
categorical_cols = df_subset.select_dtypes(include='object').columns
```

```python


for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df_subset, x=col, palette='Set3')
    plt.title(f'Diagrama de frecventa pentru {col}')
    plt.xticks(rotation=360)
    plt.tight_layout()
    plt.show()




    
```
