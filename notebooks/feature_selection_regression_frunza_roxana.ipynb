```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("../data/subset_dataset_frunza_roxana.csv")
```

```python
selected_features = ['OverallQual', 'YearBuilt', 'YearRemodAdd','MasVnrArea',]
```

```python
## Scatter plot 2D

for feature in selected_features:
    plt.figure()
    plt.scatter(data[feature], data['SalePrice'])
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(f'SalePrice vs {feature}')
    plt.show()
```

    OverallQual is the most important feature among the selected ones, then by YearBuilt and YearRemodAdd as relevant, while MasVnrArea appears to have the weakest impact on SalePrice.

```python
## Heatmap de corelatii

import plotly.express as px
corr_matrix = data[selected_features].corr()

fig = px.imshow(
    corr_matrix,
    text_auto=True,
    title="Correlation Heatmap"
)
fig.show()
```

The heatmap shows that OverallQual has the strongest positive correlation with SalePrice, meaning the overall quality of the house is the most influential predictor of price among the selected features.

YearBuilt and YearRemodAdd also show moderate positive correlations, indicating that newer or recently renovated homes tend to have higher prices, but their influence is weaker compared to OverallQual.

MasVnrArea has the weakest correlation with SalePrice, suggesting that the masonry veneer area does not significantly affect the selling price on its own.

```python
## Scatter matrix (SPLOM)

fig = px.scatter_matrix(
    data,
    dimensions=selected_features,
    title="Scatter Matrix (SPLOM) - Selected Features and SalePrice",
    color='SalePrice' 
)
fig.show()
```

    OverallQual shows the strongest and most linear relationship with SalePrice

```python
## Surface 3D

fig = px.scatter_3d(
    data,
    x='OverallQual',
    y='YearBuilt',
    z='SalePrice',
    title="3D Scatter Surface — SalePrice vs OverallQual & YearBuilt"
)

fig.show()
```

    OverallQual remains the strongest and most dominant predictor of SalePrice

```python
## Density contour / 2D heatmap

fig = px.density_contour(
    data,
    x="OverallQual",
    y="SalePrice",
    title="2D Density Contour — SalePrice vs OverallQual"
)
fig.show()

fig = px.density_heatmap(
    data,
    x="OverallQual",
    y="SalePrice",
    title="2D Heatmap — SalePrice vs OverallQual"
)
fig.show()
```

    The density contour and heatmap show that SalePrice increases clearly as OverallQual increases.

    OverallQual again shows the strongest concentration pattern with SalePrice

```python
## 3D scatter plot

fig = px.scatter_3d(
    data,
    x='OverallQual',
    y='YearBuilt',
    z='SalePrice',
    title="3D Scatter — SalePrice vs OverallQual & YearBuilt"
)

fig.show()
```

    The 3D scatter visualization confirms that OverallQual has the strongest relationship with SalePrice
    
    OverallQual is the most important feature because it shows a very strong and linear positive relationship with SalePrice across all visualizations


    YearBuilt and YearRemodAdd show a moderate increasing trend and MasVnrArea shows only a weak and scattered relationship, so it is less important

    Conclusion

    Based on the visual analysis from this lab, I conclude that the selected variables, especially OverallQual, followed by YearBuilt / YearRemodAdd — are significant and appropriate for a regression model. They show consistent positive trends with SalePrice.
