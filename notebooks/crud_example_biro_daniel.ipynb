# Post - Trimite spre server

```python
new_user = {"name": "Alex", "email": "alex@example.com"}
r = requests.post(BASE_URL, json=new_user)
print("Status:", r.status_code)
print("Raspuns:", r.json())
```

# UPDATE - PUT (actualizeaza complet un utilizator)

```python
update_data = {"name": "Alex Updated", "email": "alexu@example.com"}
r = requests.put(BASE_URL + "/1", json=update_data)
print("Status:", r.status_code)
print("Raspuns:", r.json())
```

# DELETE - DELETE (sterge un utilizator)

```python
r = requests.delete(BASE_URL + "/1")
print("Status:", r.status_code)
```
