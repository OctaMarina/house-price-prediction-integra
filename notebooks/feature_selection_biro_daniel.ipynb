```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/subdataset_Biro_Daniel.csv')

df
```

```python
px.scatter(df, x='WoodDeckSF', y='SalePrice', title='SaleCondition_Partial vs SalePrice')
```

```python
fig1 = px.violin(df, y='SalePrice', x='YrSold', title='YrSold vs SalePrice')
fig2 = px.violin(df, y='SalePrice', x='SaleCondition', title='MoSold vs SalePrice')
fig1.show()
fig2.show()
```

```python
num_df = df.select_dtypes(include='number')
corr = num_df.corr()['SalePrice'].sort_values(ascending=False)
fig = px.imshow(num_df.corr(), text_auto=True, title='Correlation Matrix (numeric columns)')
fig.show()
```

```python
px.scatter_matrix(df, dimensions=num_df.columns.tolist(), color='SalePrice')
```
