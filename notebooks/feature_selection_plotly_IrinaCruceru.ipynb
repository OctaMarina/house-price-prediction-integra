**Selectia celor mai potrivite variabile (feature-uri) pentru regresie**

```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('house_prices_subset_IrinaCruceru.csv')
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
numeric_cols = df.select_dtypes(include=['int64']).columns
corr = df[numeric_cols].corr()

fig = px.imshow(
    corr,
    text_auto=True,
    color_continuous_scale='RdBu_r',
    title="Heatmap of Correlations Between Numeric Variables"
)
fig.show()
```

Sale price seems to have a strong linear selationship with the following:
* GrLivArea	~0.71
* TotalBsmtSF	~0.64
* 1stFlrSF	~0.62

```python
fig = px.scatter(
    df,
    x='TotalBsmtSF',
    y='SalePrice',
    trendline='ols',
    title='SalePrice vs TotalBsmtSF'
)
fig.show()
```

```python
fig = px.scatter(
    df,
    x='1stFlrSF',
    y='SalePrice',
    trendline='ols',
    title='SalePrice vs 1stFlrSF '
)
fig.show()
```

```python
fig = px.scatter(
    df,
    x='GrLivArea',
    y='SalePrice',
    trendline='ols',
    title='SalePrice vs GrLivArea'
)
fig.show()
```

As per the graph the linear relationship is visible, although there is more variation for higher values.

```python
fig = px.scatter_3d(
    df,
    x='GrLivArea',
    y='TotalBsmtSF',
    z='SalePrice',
    color='SalePrice',
    color_continuous_scale='Viridis',
    title='3D Scatter: SalePrice vs GrLivArea and TotalBsmtSF'
)
fig.show()
```

For the same GrLivArea, homes with larger basements often have higher prices.

For the same TotalBsmtSF, homes with more above-ground space also have higher prices.

This suggests that both variables contribute independently to price prediction.

```python
fig = px.box(
    df,
    x='Foundation',
    y='SalePrice',
    color='Foundation',
    title='Distribution of SalePrice by Foundation Type'
)
fig.show()
```

PConc clearly more expensive than others

As basement quality improves, the SalePrice tends to increase.

```python
fig = px.density_contour(
    df,
    x='GrLivArea',
    y='SalePrice',
    title='Density Contour: SalePrice vs GrLivArea'
)
fig.update_traces(contours_coloring="fill", contours_showlabels=True)
fig.show()
```

**Final proposed variables**
* GrLivArea - Consistently the strongest positive linear relationship with SalePrice
* TotalBsmtSF - Strong positive correlation
* 1stFlrSF - Moderate-to-strong correlation, adds information about main-floor living space
* BsmtQual - Clear trend, higher quality basements results in higher prices
* CentralAir - Another clear trend, houses with central air are more expensive
* Foundation - PConc clearly more expensive than others

**Conclusion**

The analysis shows that both house size and construction quality should be significant predictors of SalePrice.

Among numerical variables, GrLivArea, TotalBsmtSF, and 1stFlrSF exhibit the strongest positive correlations, indicating that larger homes consistently sell for higher prices.

From the categorical features, Foundation, BsmtQual and CentralAir display clear price differences between categories, reflecting the impact of quality and modernization.

Although some of these features were less influential in the Decision Tree model (CentralAir), they remain important for linear regression.
