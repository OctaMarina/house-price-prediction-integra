```python
#reading the second dataset
hp=pd.read_csv("house_prices.csv")
```

```python
from sklearn.model_selection import train_test_split
```

```python
# test/train split
train_df, test_df = train_test_split(hp, test_size=0.2, random_state=42)
```

```python
#selecting the columns form set 3
cols_to_keep = [
    'Foundation', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinSF1',
    'BsmtFinType2', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', 'Heating', 'HeatingQC',
    'CentralAir', 'Electrical', '1stFlrSF', '2ndFlrSF', 'LowQualFinSF', 'GrLivArea', 'SalePrice'
]
#creating the subset
df_subset = train_df[cols_to_keep]
```
