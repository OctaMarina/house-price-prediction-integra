```python
import pandas as pd
df = pd.read_csv("house_prices.csv")
```

```python
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
df_subset = train_df[['WoodDeckSF','OpenPorchSF','EnclosedPorch','3SsnPorch','ScreenPorch','PoolArea','PoolQC','Fence','MiscFeature','MiscVal','MoSold','YrSold','SaleType','SaleCondition','SalePrice']]
df_subset
```

```python
total_nulls_per_column = df_subset.isna().sum()
total_nulls_per_column
```

```python
total_nulls = total_nulls_per_column.sum()
total_nulls
```

```python
import matplotlib.pyplot as plt
plt.hist(df_subset['SalePrice'], bins=80)
plt.show()
```

```python
numeric_columns = df_subset.select_dtypes(include='number').columns.tolist()
for column in numeric_columns:
  plt.hist(df_subset[column], bins=80)
  plt.title(column)
  plt.show()
```

```python
for column in numeric_columns:
  plt.boxplot(df_subset[column])
  plt.title(column)
  plt.show()
```

```python
import seaborn as sns

categorical_columns = df_subset.select_dtypes(include=[object]).columns.tolist()
for columns in categorical_columns:
  plt.figure(figsize=(10,5))
  sns.countplot(x=columns, data=df_subset)
  plt.xticks(rotation=45)
  plt.show()
```
