```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = "../data/subdataset_Biro_Daniel.csv"
df = pd.read_csv(path)
```

```python
# feature 1 - total outdoor area
df["TotalOutdoorArea"] = df["WoodDeckSF"] + df["OpenPorchSF"] + df["EnclosedPorch"] + df["3SsnPorch"] + df["ScreenPorch"] + df["PoolArea"]
```

```python
# feature 2 - YearsSinceSold
df["YearsSinceSold"] = 2025 - df["YrSold"]
```

```python
sale_price = df["SalePrice"]
toa = df["TotalOutdoorArea"]
yss = df["YearsSinceSold"]

sns.scatterplot(x=toa, y=sale_price)
plt.show()
sns.scatterplot(x=yss, y=sale_price)
```

TotalOutdoorArea has a very weak correlation with SalePrice. This is partly due to the fact that a lot of 0's are present in the dataset. In general an upwards trend can be noticed, but no clear relation. 

YearsSinceSold has no direct correlation with SalePrice.

```python
correlations = df[["TotalOutdoorArea", "YearsSinceSold", "SalePrice"]].corr()["SalePrice"]
correlations
```
